```python
# 개요
# ==
# 폐렴 환자 구분을 위한 ResNet 모델 구현 및 학습

# 목차
# 1. 데이터
#   1. 데이터 분석
#   2. 정상 이미지와 폐렴 환자 이미지 구분
#   3. 데이터 라벨링
#   4. train & val & test 데이터 생성
#   5. 데이터 변형
# 2. Residual block 생성
# 3. ResNet 모델 구현
# 4. 모델 컴파일
# 5. recall을 강조하기 위한 가중치 조절
# 6. 모델 학습
# 7. 결과 확인
```

```python
import os, re
import random, math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
# 필요한 변수 생성
AUTOTUNE = tf.data.experimental.AUTOTUNE

IMAGE_SIZE = [180, 180]

ROOT_PATH = os.path.join(os.getenv('HOME'), 'aiffel')
TRAIN_PATH = ROOT_PATH + '/chest_xray/data/train/*/*'
VAL_PATH = ROOT_PATH + '/chest_xray/data/val/*/*'
TEST_PATH = ROOT_PATH + '/chest_xray/data/test/*/*'

BATCH_SIZE = 32
EPOCHS = 10

print(ROOT_PATH)
```

```python
# 데이터 불러오기
train_filenames = tf.io.gfile.glob(TRAIN_PATH)
test_filenames = tf.io.gfile.glob(TEST_PATH)
val_filenames = tf.io.gfile.glob(VAL_PATH)

print(len(train_filenames))
print(len(test_filenames))
print(len(val_filenames))
```

```python
# val 개수가 적어 train과 val 데이터를 합쳐 8:2로 분할
filenames = tf.io.gfile.glob(TRAIN_PATH)
filenames.extend(tf.io.gfile.glob(VAL_PATH))

train_size = math.floor(len(filenames)*0.8)
random.seed(8)
random.shuffle(filenames)
train_filenames = filenames[:train_size]
val_filenames = filenames[train_size:]

print(len(train_filenames))
print(len(val_filenames))
```

```python
# 정상 이미지와 폐렴 이미지 구분
print(f'Normal image path\n{filenames[0]}')
print(f'Pneumonia image path\n{filenames[2000]}')
```

```python
# 정상 이미지와 폐렴 이미지 개수 확인
COUNT_NORMAL = len([filename for filename in train_filenames if "NORMAL" in filename])
print(f"Normal images count in training set: {COUNT_NORMAL}")


COUNT_PNEUMONIA = len([filename for filename in train_filenames if "PNEUMONIA" in filename])
print(f"Pneumonia images count in training set: {COUNT_PNEUMONIA}")
```

```python
# train 데이터와 val 데이터 변수에 저장
train_list_ds = tf.data.Dataset.from_tensor_slices(train_filenames)
val_list_ds = tf.data.Dataset.from_tensor_slices(val_filenames)
```

```python
# train 데이터, val 데이터 개수 확인
TRAIN_IMG_COUNT = tf.data.experimental.cardinality(train_list_ds).numpy()
print(f"Training images count: {TRAIN_IMG_COUNT}")

VAL_IMG_COUNT = tf.data.experimental.cardinality(val_list_ds).numpy()
print(f"Validating images count: {VAL_IMG_COUNT}")
```

```python
# 파일 경로에 있는 NORMAL과 PNEUMONIA 으로 라벨 데이터 생성
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2] == "PNEUMONIA"   # 폐렴이면 양성(True), 노말이면 음성(False)
```

```python
# 이미지를 float으로 변경
def decode_img(img):
    img = tf.image.decode_jpeg(img, channels=3) 
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.image.resize(img, IMAGE_SIZE) 
    return img

def process_path(file_path):
    label = get_label(file_path) 
    img = tf.io.read_file(file_path) 
    img = decode_img(img)
    return img, label
```

```python
# train 데이터와 val 데이터 생성
train_ds = train_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
val_ds = val_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
```

```python
# 이미지 확인
for image, label in train_ds.take(1):
    print("Image shape: ", image.numpy().shape)
    print("Label: ", label.numpy())
```

```python
# test 데이터 생성
test_list_ds = tf.data.Dataset.list_files(TEST_PATH)
TEST_IMAGE_COUNT = tf.data.experimental.cardinality(test_list_ds).numpy()
test_ds = test_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE)

print(TEST_IMAGE_COUNT)
```

```python
def prepare_for_training(ds, shuffle_buffer_size=1000):
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

train_ds = prepare_for_training(train_ds)
val_ds = prepare_for_training(val_ds)
```

```python
# 데이터 시각화
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(BATCH_SIZE):
        ax = plt.subplot(4,math.ceil(BATCH_SIZE/4),n+1)
        plt.imshow(image_batch[n])
        if label_batch[n]:
            plt.title("PNEUMONIA")
        else:
            plt.title("NORMAL")
        plt.axis("off")


image_batch, label_batch = next(iter(train_ds))
show_batch(image_batch.numpy(), label_batch.numpy())
```

```python
# conv block 함수
def conv_block(filters):
    block = tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ])
    
    return block
```

```python
# dense block 함수
def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    
    return block
```

```python
# 모델 생성
def build_model():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
        
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPool2D(),
        
        conv_block(32),
        conv_block(64),
        
        conv_block(128),
        tf.keras.layers.Dropout(0.2),
        
        conv_block(256),
        tf.keras.layers.Dropout(0.2),
        
        tf.keras.layers.Flatten(),
        dense_block(512, 0.7),
        dense_block(128, 0.5),
        dense_block(64, 0.3),
        
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    
    return model
```

```python
# 가중치 처리
weight_for_0 = (1 / COUNT_NORMAL)*(TRAIN_IMG_COUNT)/2.0 
weight_for_1 = (1 / COUNT_PNEUMONIA)*(TRAIN_IMG_COUNT)/2.0

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
```

```python
# 모델 훈련
with tf.device('/GPU:0'):
    model = build_model()

    METRICS = [
        'accuracy',
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
    
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=METRICS
    )
with tf.device('/GPU:0'):
    history = model.fit(
        train_ds,
        steps_per_epoch=TRAIN_IMG_COUNT // BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=val_ds,
        validation_steps=VAL_IMG_COUNT // BATCH_SIZE,
        class_weight=class_weight,
    )
```

```python
# 결과 확인
fig, ax = plt.subplots(1, 4, figsize=(20, 3))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
    
loss, accuracy, precision, recall = model.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
```

```python
# 기존 데이터 변형
def augment(image,label):
    image = tf.image.random_flip_left_right(image)
    return image,label

def prepare_for_training(ds, shuffle_buffer_size=1000, repeat=True):
    ds = ds.map(
        augment,
        num_parallel_calls=AUTOTUNE
    )
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    if repeat:
        ds = ds.repeat()  # 훈련 데이터에만 repeat 적용
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

train_ds = prepare_for_training(train_ds, repeat=True)
val_ds = prepare_for_training(val_ds, repeat=False)
```

```python
# 데이터 시각화
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(BATCH_SIZE):
        ax = plt.subplot(4,math.ceil(BATCH_SIZE/4),n+1)
        plt.imshow(image_batch[n])
        if label_batch[n]:
            plt.title("PNEUMONIA")
        else:
            plt.title("NORMAL")
        plt.axis("off")


image_batch, label_batch = next(iter(train_ds))
show_batch(image_batch.numpy(), label_batch.numpy())
```

```python
# Residual block
# GPT의 도움을 받아 작성했습니다.
from tensorflow.keras.layers import Layer, Input, Conv2D, BatchNormalization, ReLU, Add, MaxPooling2D, GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

class ResidualBlock(Layer):
    def __init__(self, filters, kernel_size=3, stride=1, use_conv_shortcut=False): # 레이어의 초기화
        super(ResidualBlock, self).__init__()
        self.use_conv_shortcut = use_conv_shortcut
        self.stride = stride
        self.filters = filters

        # Main path 컴포넌트
        self.conv1 = Conv2D(filters, kernel_size, strides=stride, padding="same", use_bias=False) # 첫 번째 2D 컨볼루션 레이어
        self.bn1 = BatchNormalization()
        self.relu1 = ReLU()

        self.conv2 = Conv2D(filters, kernel_size, strides=1, padding="same", use_bias=False) # 두 번째 2D 컨벌루션 레이어
        self.bn2 = BatchNormalization()

        # Shortcut path
        if self.use_conv_shortcut or self.stride != 1: # 쇼트컷 경로가 필요한 경우 1X1 컨볼루션 레이어 사용
            self.conv_shortcut = Conv2D(filters, 1, strides=stride, padding="same", use_bias=False)
            self.bn_shortcut = BatchNormalization()

    def call(self, inputs, training=False): # 레이어의 실행 로직 담당
        x = inputs

        # 입력 데이터에 주 경로의 컨볼루션, 배치 정규화, 활성화 함수 적용
        out = self.conv1(x)
        out = self.bn1(out, training=training)
        out = self.relu1(out)

        out = self.conv2(out)
        out = self.bn2(out, training=training)

        if self.use_conv_shortcut or self.stride != 1: # 쇼트컷 경로의 조건에 부합하는 경우 x에 쇼트컷 컨볼루션과 배치 정규화 적용
            x = self.conv_shortcut(x)
            x = self.bn_shortcut(x, training=training)

        out = Add()([x, out]) # 주 경로와 쇼트컷 경로의 결과를 합침
        out = ReLU()(out) # 최종 ReLU 활성화 함수 적용
        return out
```

```python
# resnet 구현
# GPT의 도움을 받아 작성
def ResNet18(input_shape=(180, 180, 3), classes=2): # 180x180 크기의 3채널(컬러) 이미지이고, 출력 클래스는 2개
    inputs = Input(shape=input_shape)
    
    x = Conv2D(64, 7, strides=2, padding="same", use_bias=False)(inputs) # 첫 번째 컨볼루션 레이어
    x = BatchNormalization()(x) # 컨볼루션 레이어 후 배치 정규화를 적용하여 입력 데이터의 분포를 정규화
    x = ReLU()(x) # 배치 정규화 후 비선형 활성화 함수 ReLU를 적용
    x = MaxPooling2D(3, strides=2, padding="same")(x) # 풀링 레이어를 통해 특성 맵의 크기를 줄이고 중요한 정보 유지

    # 호출은 다른 수의 필터와 설정으로 네트워크의 깊이에 따라 적용
    x = ResidualBlock(64)(x)
    x = ResidualBlock(64)(x)

    # use_conv_shortcut=True 쇼트컷 연결에서 차원의 불일치를 해결하기 위해 1x1 컨볼루션을 사용
    x = ResidualBlock(128, stride=2, use_conv_shortcut=True)(x)
    x = ResidualBlock(128)(x)

    x = ResidualBlock(256, stride=2, use_conv_shortcut=True)(x)
    x = ResidualBlock(256)(x)

    x = ResidualBlock(512, stride=2, use_conv_shortcut=True)(x)
    x = ResidualBlock(512)(x)

    x = GlobalAveragePooling2D()(x) # 각 특성 맵의 전역 평균을 계산하여 각 채널별로 단일 숫자를 출력
    outputs = Dense(1, activation='sigmoid')(x)  # softmax로 모델 학습이 안되서 sigmoid로 변경

    model = Model(inputs, outputs)
    return model

# 모델 생성
model = ResNet18()
model.summary()
```

```python
# 모델 컴파일
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K

K.clear_session() # 모델의 상태 초기화
model = ResNet18()
model.compile(optimizer=Adam(learning_rate=0.0001, clipnorm=1.0), # 안정적인 학습을 위해 학습률을 매우 낮게 설정 및 그라디언트 폭발 문제 방지
              loss='binary_crossentropy', # 이진 분류 문제이기 때문에 사용
              metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
```

```python
# recall을 강조하기 위한 가중치 조절
weight_for_0 = (1 / COUNT_NORMAL) * (TRAIN_IMG_COUNT) / 2.0 
weight_for_1 = (1 / COUNT_PNEUMONIA) * (TRAIN_IMG_COUNT) * 2.0  # 폐렴에 더 높은 가중치

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
```

```python
# 모델 학습
history = model.fit(
    train_ds,
    steps_per_epoch=TRAIN_IMG_COUNT // BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=val_ds,
    validation_steps=VAL_IMG_COUNT // BATCH_SIZE,
    class_weight=class_weight,
)
```

```python
# 결과 확인 (폐렴에 더 높은 가중치 부여)
fig, ax = plt.subplots(1, 4, figsize=(20, 3))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
    
loss, accuracy, precision, recall = model.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
# precision : 훈련 데이터에 대한 정밀도는 높지만 검증 데이터에 대해서는 정밀도가 살짝 낮고 변동성이 큼
# recall : 재현율은 매우 높지만 정밀도가 낮기 때문에 positive 클래스를 잘 못 분류하는 경우도 많을 수 있음
# accuracy : 훈련 데이터에 비해 검증 데이터에서 낮게 나타나고 있음
# loss : 손실 그래프에서는 훈련 손실이 지속적으로 감소하는 반면, 검증 손실은 일정한 에폭에서 급증, 특정 시점에서 과적합이 시작할 수 있음
```

```python
# 결과 확인 (같은 가중치 부여)
fig, ax = plt.subplots(1, 4, figsize=(20, 3))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
    
loss, accuracy, precision, recall = model.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
# precision : 앞서 실행했던 모델과 비교했을 때 더 높은 정밀도를 가지고 있음
# recall : 앞서 실행했던 모델이 재현율이 좋지만 정밀도가 떨어지기 때문에 이번 모델의 재현율이 더 믿을 수 있음
# accuracy : 앞서 실행한 모델보다 정확도가 미세하게 높음
# loss : 앞서 실행한 모델보다 손실이 더 낮음
```
