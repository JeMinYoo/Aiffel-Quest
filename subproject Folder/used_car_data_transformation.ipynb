```python
import pandas as pd
import numpy as np
```

```python
car_data = pd.read_csv('/aiffel/data/cars.csv')
brand_data = pd.read_csv('/aiffel/data/brand.csv')
```

```python
car_data.head()
```

```python
brand_data.head()
```

```python
# car_data에 있는 title의 데이터 값을 brand_data와 맞추기 위해 ' ' 변경
car_data['title'] = car_data['title'].apply(lambda x: x.split()[0].lower())
```

```python
# 두 데이터프레임을 title을 기준으로 병합
merged_df = pd.merge(car_data, brand_data, on='title', how='left')
```

```python
merged_df
```

```python
merged_df['Engine'].unique()
```

```python
merged_df['Emission Class'].unique()
```

```python
merged_df['Engine'] = pd.Categorical(merged_df['Engine']).codes
merged_df['Emission Class'] = pd.Categorical(merged_df['Emission Class']).codes
```

```python
merged_df
```

```python
merged_df.isnull().sum()
```

```python
# Service history 변수의 결측치를 Unknown으로 채우기
merged_df['Service history'].fillna('Unknown', inplace=True)
```

```python
merged_df
```

```python
# 3개 이상의 결측치가 있는 행 제거
threshold = 3
merged_df = merged_df.dropna(thresh=merged_df.shape[1] - threshold + 1)
```

```python
merged_df.isnull().sum()
```

```python
merged_df
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
```

```python
ss = StandardScaler()
rs = RobustScaler()
mm = MinMaxScaler()
```

```python
# 문자열 값을 숫자형으로 변환
merged_df = pd.get_dummies(merged_df, drop_first=True)
```

```python
ss_merged_df = pd.DataFrame(ss.fit_transform(merged_df), columns=merged_df.columns)
```
