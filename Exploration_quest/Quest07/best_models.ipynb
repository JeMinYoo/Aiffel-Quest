```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, BatchNormalization, Flatten, Dense, Dropout

# Sequential 모델 초기화
model = Sequential()

# 첫 번째 컨볼루션 레이어 + 평균 풀링 + 배치 정규화
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(AveragePooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())

# 두 번째 컨볼루션 레이어 + 평균 풀링 + 배치 정규화
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(AveragePooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())

# 플래튼 레이어
model.add(Flatten())

# 첫 번째 Dense 레이어
model.add(Dense(256, activation='relu'))

# 드롭아웃 레이어
model.add(Dropout(0.5))

# 두 번째 Dense 레이어 (출력 레이어)
model.add(Dense(10, activation='softmax'))

# 모델 요약 출력
model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10

# CIFAR-10 데이터셋 불러오기
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# 데이터 정규화 (0~1 범위로)
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 모델 컴파일 (최적화 방법, 손실 함수, 평가 지표 설정)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델 학습 (시간이 걸릴 수 있음)
model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test))

# 예측 클래스 얻기
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # 예측한 클래스 인덱스
y_test_classes = y_test.reshape(-1)  # 실제 클래스 인덱스

# 잘못된 예측 찾기
incorrect_indices = np.where(y_pred_classes != y_test_classes)[0]

# 잘못된 예측 이미지 몇 개 출력
num_errors_to_display = 10  # 출력할 이미지 수

for i in range(num_errors_to_display):
    index = incorrect_indices[i]
    plt.imshow(x_test[index])
    plt.title(f"실제 레이블: {y_test_classes[index]}, 예측: {y_pred_classes[index]}")
    plt.show()
```

```python
import kerastuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, BatchNormalization, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

def build_model(hp):
    model = Sequential()

    # 첫 번째 컨볼루션 레이어 + 평균 풀링 + 배치 정규화
    model.add(Conv2D(hp.Int('conv1_filters', min_value=32, max_value=64, step=16),
                     (3, 3), activation='relu', input_shape=(32, 32, 3)))
    model.add(AveragePooling2D(pool_size=(2, 2)))
    model.add(BatchNormalization())

    # 두 번째 컨볼루션 레이어 + 평균 풀링 + 배치 정규화
    model.add(Conv2D(hp.Int('conv2_filters', min_value=64, max_value=128, step=32),
                     (3, 3), activation='relu'))
    model.add(AveragePooling2D(pool_size=(2, 2)))
    model.add(BatchNormalization())

    # 플래튼 레이어
    model.add(Flatten())

    # 첫 번째 Dense 레이어
    model.add(Dense(hp.Int('dense_units', min_value=128, max_value=512, step=128), activation='relu'))

    # 드롭아웃 레이어
    model.add(Dropout(hp.Float('dropout_rate', min_value=0.3, max_value=0.7, step=0.1)))

    # 두 번째 Dense 레이어 (출력 레이어)
    model.add(Dense(10, activation='softmax'))

    # 컴파일 옵션
    model.compile(optimizer=Adam(hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='LOG')),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model
```

```python
from kerastuner.tuners import RandomSearch

# 튜너 설정
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,  # 시도할 실험의 수
    executions_per_trial=2,  # 각 조합당 실행 횟수
    directory='my_dir',
    project_name='cifar10_tuning'
)

# 조기 종료를 위한 콜백 추가 (선택사항)
from tensorflow.keras.callbacks import EarlyStopping

stop_early = EarlyStopping(monitor='val_loss', patience=5)

# 튜닝 시작
tuner.search(x_train, y_train, epochs=10, validation_data=(x_test, y_test_classes), callbacks=[stop_early])
```

```python
# 최적의 하이퍼파라미터로 모델 학습
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

model = build_model(best_hps)

history = model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test_classes))
```

```python
# 모델을 SavedModel 형식으로 저장
model.save('saved_model/my_cifar10_model')
```

```python
# 모델 평가 (테스트 정확도 체크)
test_loss, test_accuracy = model.evaluate(x_test, y_test_classes, verbose=2)

print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
